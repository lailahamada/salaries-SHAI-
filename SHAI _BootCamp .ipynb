# Q 3 

```python
# Handling missing data  # dropna( axis=1 ) is a function used to drop columns with missing values data contain Nana so , we can drop it 
df = df.dropna(axis=1)
df.isna().sum()
```

# Q 4

```python
plt.figure(figsize=(8,6))
sns.histplot(df.TotalPay,bins=40)
plt.xlabel("Frequency")
plt.ylabel("TotalPay")
plt.title("Distribution of TotalPay")
plt.show()
```

```python
df['JobTitle'].value_counts()
```

```python
plt.figure(figsize=(5,5))

plt.pie(df['JobTitle'].value_counts()),

plt.show()
```

# Q 5

```python
grouped_by_job_title = df.groupby('JobTitle')['TotalPay'].mean()
print("Average Salary by Job Title: ",grouped_by_job_title)
```

```python
grouped_by_TotalPayBenefits= df.groupby('TotalPayBenefits')['TotalPay'].mean()
print("Average Salary by TotalPayBenefits: ",grouped_by_job_title)
```
